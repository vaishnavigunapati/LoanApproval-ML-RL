```python
# =============================
# 4_Comparison_DL_vs_RL.ipynb
# =============================

# -----------------------------
# 1️⃣ IMPORT LIBRARIES
# -----------------------------
import numpy as np
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix

# -----------------------------
# 2️⃣ LOAD / DEFINE DL METRICS


# DL metrics
dl_auc = 0.7002
dl_f1 = 0.4129

# Confusion matrix from DL
dl_cm = np.array([[122925, 80947],
                  [15548, 33937]])

# -----------------------------
# 3️⃣ LOAD / DEFINE RL METRICS
# -----------------------------
# Example: replace with your actual RL policy value
rl_policy_value = 0.1046

# RL baseline: always approve / always deny

rl_baseline_approve = 0.08
rl_baseline_deny = 0.0

# -----------------------------
# 4️⃣ COMPARE DL vs RL
# -----------------------------
print("===== DL vs RL COMPARISON =====\n")
print("DL Model Metrics:")
print(f"AUC: {dl_auc:.4f}")
print(f"F1-Score: {dl_f1:.4f}")
print("Confusion Matrix:\n", dl_cm)

print("\nRL Agent Metrics:")
print(f"Policy Value (DiscreteBC): {rl_policy_value:.4f}")
print(f"Baseline Always Approve: {rl_baseline_approve:.4f}")
print(f"Baseline Always Deny: {rl_baseline_deny:.4f}")

# -----------------------------

# -----------------------------
# 6️⃣ OPTIONAL: Visual Comparison
# -----------------------------
import matplotlib.pyplot as plt

labels = ["DL AUC", "DL F1", "RL Policy Value", "RL Baseline Approve", "RL Baseline Deny"]
values = [dl_auc, dl_f1, rl_policy_value, rl_baseline_approve, rl_baseline_deny]

plt.figure(figsize=(10,5))
plt.bar(labels, values, color=['skyblue','lightgreen','salmon','orange','gray'])
plt.title("DL vs RL Metrics Comparison")
plt.ylabel("Value")
plt.show()
```
